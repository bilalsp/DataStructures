# <Center> <HR> <font color="blue"> Segregate 0s and 1s in an array </font> <HR> </Center>
   
   Let S: $(a_1, a_2, ..., a_n)$ be the sequence of numbers. Rearrange the sequence, $S = (S_{0},S_{1})$ such that
   $$ \text{ x = 0 } \forall x \in S_{0} $$
   $$ \text{ y = 1 } \forall y \in S_{1} $$

## 1. Using Partioning Approach
    Time complexity : O(n) || Auxiliary Space : O(1)

```python
def rearrange_1(arr, n):
    
    zero_index = 0
    for i in range(n):
        if arr[i] == 0:
            arr[zero_index], arr[i] = arr[i], arr[zero_index]
            zero_index += 1
```

## 2. Using two indexes to traverse
    Time complexity : O(n) || Auxiliary Space : O(1)

```python
def rearrange_2(arr, n):
    
    left, right = 0, n-1
    
    while left < right:
        
        while arr[left] == 0 and left < right:
            left += 1
            
        while arr[right] == 1 and left < right:
            right -= 1
            
        if left < right:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
```

## Testing

```python
n = int(input())
arr = list(map(int, input().strip().split()))

rearrange_1(arr, n)

print(*arr)
```

## Comparison of Different Approaches

```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt 
%matplotlib inline


n_execution = 25
random_Size = lambda x: 100*x*np.random.randint(1,100)
arr_list = [np.concatenate([np.zeros(random_Size(i)), np.ones(random_Size(i))]) for i in range(1, n_execution+1)]

for fun in ["rearrange_1", "rearrange_2"]:
    
    times, arr_size = [], []
    for arr in arr_list:
        random.shuffle(arr)
        n = len(arr)
        
        s = time.time()
        eval(fun)(arr, n)
        e = time.time()
        
        times.append(e-s)
        arr_size.append(n)
        
    plt.plot(arr_size, times, label = fun) 
    
plt.xlabel('Sequence Size') 
plt.ylabel('Run Time') 
plt.grid() 
plt.legend()
plt.show() 
```
